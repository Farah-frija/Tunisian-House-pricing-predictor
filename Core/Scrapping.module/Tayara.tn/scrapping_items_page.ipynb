# **Tayara Preprocessing**

Imports

```python
import pandas as pd
import os

file_path = r"rawData/tayara_listings.csv"

if os.path.exists(file_path):
    df1 = pd.read_csv(file_path)
    print("File loaded, shape:", df1.shape)
else:
    print("File not found at:", os.path.abspath(file_path))
    
file_path = r"rawData/tayara_immobineuf_listings.csv"

if os.path.exists(file_path):
    df2 = pd.read_csv(file_path)
    print("File loaded, shape:", df2.shape)
else:
    print("File not found at:", os.path.abspath(file_path))
    
df = pd.concat([df1, df2], ignore_index=True)
print(" Tayar's raw data size ",len(df))
print("tayara's categories")
print(df['category'].unique())
```

```python
df['neuf'] = df['category'] == "Immobilier Neuf"
```

# Les differentes catégories 

```python
df['category'].unique()
```

Notre projet se focalise seulement sur les categories 'Maisons et Villas' , 'Appartements' et 'Immobilier Neuf'

```python
df_clean=df[df.category.isin ( [ 'Maisons et Villas', 'Appartements','Immobilier Neuf'])].reset_index(drop=True)
len(df_clean)
```

```python
# Split location_time into location and time_ago

from datetime import datetime
import datetime as dt
import re
df=df_clean

df[['location', 'time_ago']] = df['location_time'].str.split(',', expand=True)
df['location'] = df['location'].str.strip()
df['time_ago'] = df['time_ago'].str.strip()
```

Ici on a l'attribut location represente la region

```python
df['location'].unique()
```

Selection des colonnes

```python
df['id']=df.index
df=df.rename(columns={
    'url': 'url',
    'title': 'titre',
    'price': 'prix',
    'location_time': 'location_time',
    'category': 'categorie',
    'seller': 'seller',
    'location': 'gouvernorat',
    'time_ago': 'published_on'
})
```

```python
#colonnes à considerer 
cols=["id","url","titre","categorie","gouvernorat",'neuf']
df=df[cols]
```
