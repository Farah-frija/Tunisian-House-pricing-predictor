# **Tayara Preprocessing**

Imports

```python
import pandas as pd
import os

file_path = r"rawData/tayara_listings_complete_maisons_appart.csv"

if os.path.exists(file_path):
    df1 = pd.read_csv(file_path)
    print("File loaded, shape:", df1.shape)
else:
    print("File not found at:", os.path.abspath(file_path))
    
file_path = r"rawData/tayara_listings_complete_neuf.csv"

if os.path.exists(file_path):
    df2 = pd.read_csv(file_path)
    print("File loaded, shape:", df2.shape)
else:
    print("File not found at:", os.path.abspath(file_path))
    
df = pd.concat([df1, df2], ignore_index=True)
print(" Tayar's raw data size ",len(df))
print("tayara's categories")
print(df['categorie'].unique())
```

```python
df['neuf'] = df['categorie'] == "immoneuf"
```

```python
# 1. Voir les IDs dupliqués
duplicates = df[df['id'].duplicated(keep=False)]  # keep=False montre toutes les occurrences
print(f"Nombre total de lignes avec IDs dupliqués: {len(duplicates)}")
```

```python
#supprimer les ids dupliqués
df= df.drop_duplicates(subset='id', keep='first')
```

```python
# 1. Voir les titres dupliqués
duplicates = df[df['titre'].duplicated(keep=False)]  # keep=False montre toutes les occurrences
print(f"Nombre total de lignes avec titres dupliqués: {len(duplicates)}")
# 2. Voir les descs dupliqués
duplicates = df[df['id'].duplicated(keep=False)]  # keep=False montre toutes les occurrences
print(f"Nombre total de lignes avec descriptions dupliqués: {len(duplicates)}")
```
